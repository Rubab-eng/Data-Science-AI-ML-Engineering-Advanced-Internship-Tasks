```python
import pandas as pd                       #Import Important Libraries
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]    #Select features for clustering
```

```python
scaler = StandardScaler()                #Feature scaling
```

```python
X_scaled = scaler.fit_transform(X)
```

```python
wcss = []      #Find the optimal number of clusters using the Elbow Method
```

```python
for k in range(1, 11):kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
```

```python
kmeans.fit(X_scaled)
```

```python
wcss.append(kmeans.inertia_)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', random_state=42)         # Apply KMeans with the chosen number of clusters (e.g., 5)
```

```python
y_kmeans = kmeans.fit_predict(X_scaled)
```

```python
df['Cluster'] = y_kmeans   #cluster labels to the original DataFrame
```

```python
plt.figure(figsize=(8, 6))      #Visualization of Clusters
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=y_kmeans, palette='Set2', s=100)
```

```python
plt.title('Customer Segments using KMeans')
```

```python
plt.xlabel('Annual Income (scaled)')
```

```python
plt.ylabel('Spending Score (scaled)')
```

```python
plt.grid(True)
```

```python
clusters_named = pd.Series(y_kmeans).map({0: 'Cluster 1', 1: 'Cluster 2', 2: 'Cluster 3', 3: 'Cluster 4', 4: 'Cluster 5'})  #using y-kmeans Clustering
```

```python
plt.figure(figsize=(8, 6))
```

```python
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=clusters_named, palette='Set2', s=100)
```

```python
plt.title('Customer Segments using KMeans')
```

```python
plt.xlabel('Annual Income (scaled)')
```

```python
plt.ylabel('Spending Score (scaled)')
```

```python
plt.grid(True)
```

```python
from sklearn.decomposition import PCA             #PCA for 2D Visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pca = PCA(n_components=2)                   #Reduce to 2 principal components
pca_components = pca.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
```

```python
sns.scatterplot(x=pca_components[:, 0], y=pca_components[:, 1], hue=y_kmeans, palette='Set2', s=100)
```

```python
plt.title('K-Means Clusters Visualized with PCA')
```

```python
plt.xlabel('PCA Component 1')
```

```python
plt.ylabel('PCA Component 2')
```

```python
from sklearn.manifold import TSNE              # t-SNE for 2D Visualization
```
